```python
#使用寝室门禁特征
feature_columns=[subsidy_feature,card_feature,score_feature,library_feature,dorm_feature,borrow_feature]

#不用寝室门禁特征
#feature_columns=[subsidy_feature,card_feature,score_feature,library_feature,borrow_feature]

#定义连接所有特征函数
def get_merge_features(features):
    for x in range(len(features)-1):
        if x==0:
            all_features=pd.merge(features[x],features[x+1],how='left',on='id')
        else:
            all_features=pd.merge(all_features,features[x+1],how='left',on='id')
            x+=1
    return all_features

#运用函数
all_features=get_merge_features(feature_columns)
```

```python
all_features.head()
```

```python
#dorm统计量

dorm_sts=dorm_feature['dorm_daycount'].describe()
dorm_IQR=dorm_sts.ix['75%']-dorm_sts.ix['25%']
#小于（下四分位数-1.5*四分位距）或大于（上四分位数+1.5*四分位距）的值可以认为是异常值
dorm_outlier=dorm_sts.ix['25%']-1.5*dorm_IQR
#这里只认为“门禁记录总天数小于21”的为异常值
dorm_outlier
```

```python
#处理dorm缺失值与异常值
#import decimal
#decimal.getcontext().prec = 100
def dorm_preprocessing(df,df1):
    #特征名
    a=[x for x in df1.columns if x not in ['id']]
    #异常数据
    df.fillna(value=0,inplace=True)
    df_outlier_dorm=df.loc[df['dorm_daycount']<=25]
    #学院编号
    b=list(df_outlier_dorm['college'].unique())
    #正常数据
    df_normal_dorm=df.loc[df['dorm_daycount']>25]
    #迭代为异常id的每个特征填充随机数（每个特征均值一个标准差范围内），分学院填充
    for j in range(len(b)):
        if j==0:
            df_outlier_college_d=df_outlier_dorm.loc[df_outlier_dorm['college']==b[j]]
            df_normal_college_d=df_normal_dorm.loc[df_normal_dorm['college']==b[j]]
            for i in range(len(a)):
                df_mean=df_normal_college_d[a[i]].mean()
                df_std=df_normal_college_d[a[i]].std()
                rand_card=np.random.uniform(df_mean - df_std, df_mean + df_std, size = len(df_outlier_college_d))
                df_outlier_college_d[a[i]]=rand_card
        else:
            df_outlier_college_j=df_outlier_dorm.loc[df_outlier_dorm['college']==b[j]]
            df_normal_college_j=df_normal_dorm.loc[df_normal_dorm['college']==b[j]]
            for i in range(len(a)):
                df_mean=df_normal_college_j[a[i]].mean()
                df_std=df_normal_college_j[a[i]].std()
                rand_card=np.random.uniform(df_mean-df_std, df_mean+df_std, size = len(df_outlier_college_j))
                df_outlier_college_j[a[i]]=rand_card
            df_outlier_college_d=pd.concat([df_outlier_college_d,df_outlier_college_j])
    df=pd.concat([df_normal_dorm,df_outlier_college_d])
    return df

#运用函数
final_feature=dorm_preprocessing(all_features,dorm_feature) 
```

```python
#card特征统计量

card_sts=card_feature['card_day_sum'].describe()
#四分位距
card_IQR=card_sts.ix['75%']-card_sts.ix['25%']
#小于（下四分位数-1.5*四分位距）或大于（上四分位数+1.5*四分位距）的值可以认为是异常值

#这里只认为“消费总天数小于137”的为异常值
card_outlier=card_sts['25%']-1.5*card_IQR
card_outlier
```

```python
#处理card缺失值与异常值
import decimal
#decimal.getcontext().prec = 100
def card_preprocessing(df,df1):
    #特征名
    a=[x for x in df1.columns if x not in ['id']]
    #异常数据
    #df.fillna(value=0,inplace=True)
    df_outlier=df.loc[df['card_day_sum']<=137]
    #学院编号
    b=list(df_outlier['college'].unique())
    #正常数据
    df_normal=df.loc[df['card_day_sum']>137]
    #迭代为异常id的每个特征填充随机数（每个特征均值一个标准差范围内），分学院填充
    for j in xrange(len(b)):
        if j==0:
            df_outlier_college=df_outlier.loc[df_outlier['college']==b[j]]
            df_normal_college=df_normal.loc[df_normal['college']==b[j]]
            for i in range(len(a)):
                #注意方差有可能比均值大，需要取绝对值
                df_mean=df_normal_college[a[i]].mean()
                df_std=df_normal_college[a[i]].std()
                rand_card=np.random.uniform(df_mean - df_std, df_mean + df_std, size = len(df_outlier_college))
                df_outlier_college[a[i]]=rand_card
        else:
            df_outlier_college_j=df_outlier.loc[df_outlier['college']==b[j]]
            df_normal_college_j=df_normal.loc[df_normal['college']==b[j]]
            for i in xrange(len(a)):
                df_mean=df_normal_college_j[a[i]].mean()
                df_std=df_normal_college_j[a[i]].std()
                rand_card=np.random.uniform(df_mean-df_std, df_mean+df_std, size = len(df_outlier_college_j))
                df_outlier_college_j[a[i]]=rand_card
            df_outlier_college=pd.concat([df_outlier_college,df_outlier_college_j])
    df=pd.concat([df_outlier_college,df_normal])
    return df

#运用函数
final_feature=card_preprocessing(final_feature,card_feature)
```

```python
#该生所在学院的获得1000，1500，2000助学人数所占的比例(测试集不用运行此项)

def get_subsidy_percent(df):
    #学院编号列表
    a=list(df.groupby('college')['id'].count().index)
    b=[0,1000,1500,2000]
    for i in range(len(b)):
        for j in range(len(a)):
                college=df[df['college']==a[j]]#学院
                subsidy_num=college.loc[college['money']==b[i]]#学院获奖为b[i]的总人数
                subsidy_id=len(subsidy_num)
                college_id=pd.Series(college['id'].values)#学院id列表集合
                subsidy_df_i_j=DataFrame([college_id]*2).T#初始化表格
                subsidy_df_i_j.columns=['id','num']
                subsidy_df_i_j['num']=subsidy_id
                subsidy_df_i_j[str(b[i])+'_percent']=np.divide.reduce([subsidy_df_i_j['num'].values,college['college_num'].values],axis=0)#学院获奖为b[j]的人数占比
                subsidy_df_i_j.drop('num',axis=1,inplace=True)
                if j==0:
                    subsidy_df=subsidy_df_i_j
                else:
                    subsidy_df=pd.concat([subsidy_df,subsidy_df_i_j])
        df=pd.merge(df,subsidy_df,how='left',on='id')
    return df

#运用函数        
final_feature=get_subsidy_percent(final_feature)
```

```python
#获得每个学院的获奖人数比例(测试集不用运行此项,结果给测试集使用)

percent_college=DataFrame(final_feature.groupby(['college','0_percent','1000_percent','1500_percent','2000_percent'])['id'].count())
percent_college.to_csv('all_feature_input/percent.csv',index=True)
percent_college=pd.read_csv('all_feature_input/percent.csv')
percent_college.drop('id',axis=1,inplace=True)
percent_college
```

```python
#连接测试集(训练集不用运行此项)

final_feature=pd.merge(final_feature,percent_college,how='left',on='college')
```

```python
#学生所在学院的消费排名（倒序）

def get_amount_rank(df):
    a=list(df.groupby('college')['id'].count().index)#学院编号列表
    for i in range(len(a)):
        college=df.loc[df['college']==a[i]]
        college_id=pd.Series(college['id'].values)
        amount_df_i=pd.DataFrame([college_id]).T#初始化表格
        amount_df_i.columns=['id']
        amount_df_i['amount_college_rank']=college['amount_mean_day'].rank(method='first',na_option='keep',ascending=False).values
        if i==0:
            amount_df=amount_df_i
        else:
            amount_df=pd.concat([amount_df,amount_df_i])
            
    df=pd.merge(df,amount_df,how='left',on='id')
    return df
      
#运用函数
final_feature=get_amount_rank(final_feature)
```

```python
#学生所在学院的消费排名与所在学院的成绩排名的乘积

final_feature['amount_score_rank']=final_feature['amount_college_rank'].multiply(final_feature['score'],fill_value=0)
```
